The second task:

* accuracy of free throws

We shall use the following approach:
* element-wise division of 2 matrix with equal dimensions: for each player we divide number of free throws by free throws attempts;
* multiplication by 100 in order to measure accuracy in %;
* let us round by 2 digits after coma 

```python
#accuracy calculation
accuracy = np.round(((matrix_ft / matrix_fta) *100), 2)

# Check the result
print("Result shape:", accuracy.shape)
print(accuracy)
```

Let us visualize accuracy per season over player:

```python
#let us create a function
def player_season_accuracy_plot(playerlist):
    #let us connect relevant accuracy, season and player
    for i, player in enumerate(playerlist):
        plt.plot(Seasons, accuracy[i], marker='s', ls='--', ms=7, label=player)  
    # pin a legend on the right side and out of the container    
    plt.legend(loc='upper right', bbox_to_anchor=(1.5, 0.5))
    plt.xticks(Seasons, rotation='vertical')

    # Add labels and title
    plt.xlabel("Season")
    plt.ylabel("Accuracy")
    plt.title("Free Throw Accuracy vs. Season, %")

    # Show the plot
    plt.show()

# make a call
player_season_accuracy_plot(Players)
```
